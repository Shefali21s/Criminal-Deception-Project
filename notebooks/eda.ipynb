```python
import pandas as pd

# Load the cleaned (not yet preprocessed) dataset
df = pd.read_csv("C:/Users/Shefali/OneDrive/Desktop/Criminal-Deception-Project/outputs/combined_clean.csv")

# Preview
df.head()
```

```python
# Check shape
print("Shape:", df.shape)

# Show column names
print("Columns:", df.columns.tolist())

# Check for null values
print("\nMissing values:\n", df.isnull().sum())

# Data types
df.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Label count
sns.countplot(data=df, x="label", hue="source")
plt.title("Label Distribution by Source")
plt.xlabel("Label (0 = Deceptive, 1 = Truthful)")
plt.ylabel("Count")
plt.show()
```

```python

# Add text length column
df["text_length"] = df["text"].apply(len)

# Plot distribution
sns.histplot(df["text_length"], bins=50, kde=True)
plt.title("Distribution of Text Lengths")
plt.xlabel("Number of Characters")
plt.show()
```

```python
print("🟥 Sample Deceptive Statements:\n")
print(df[df['label'] == 0]['text'].sample(3).values)

print("\n🟩 Sample Truthful Statements:\n")
print(df[df['label'] == 1]['text'].sample(3).values)
```
